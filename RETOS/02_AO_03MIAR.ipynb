# Medir el tiempo de ejecución

Medir el tiempo de ejecución de un programa es una tarea muy importante, ya que programar no sólo consiste en crear código que funcione, sino código que pueda ser ejecutado en un tiempo razonable.

La librería **time** nos permite medir el tiempo de ejecución en Python, utilizando el método  `time()`.

```python
# Medida del tiempo de ejecucion con time
import time
inicio = time.time()

# Código a medir
time.sleep(10)
# -------------

fin = time.time()
print(fin-inicio) # 1.0005340576171875
```

## Implementa  una solución para el problema del cambio de moneda en un sistema monetario con las siguientes monedas: {1, 2, 5, 10, 20 , 50}

```python
# Problema de cambio de monedas
import numpy as np
def cambio_de_moneda_voraz(monedas_disponibles, C):
    # Checkeamos si existe el valor unitario en el conjunto de monedas
    if 1 not in monedas_disponibles:
        raise ValueError("Debe existir el valor unitario (1) en el conjunto de monedas")

    # Ordenamos las monedas de mayor a menor
    monedas_disponibles.sort(reverse=True)

    # Lista para almacenar el número de monedas utilizadas
    num_monedas_utilizadas = np.zeros(len(monedas_disponibles), dtype=int)

    # Iteramos sobre cada moneda
    for i in range(len(monedas_disponibles)):
        # Mientras sea mayor de 0 el monto, continuamos
        if C > 0:
            # Calculamos el número de monedas maximas para cada valor 
            num_monedas_utilizadas[i] = C // monedas_disponibles[i]
            C -= num_monedas_utilizadas[i] * monedas_disponibles[i]

    cant_monedas = dict(zip(monedas_disponibles, num_monedas_utilizadas))
    
    # Devolvemos un diccionario que indica con su key el valor de la maneda
    # y en su value indica el num de esa moneda que se devuelve
    return cant_monedas
```

```python
# Testear Problema
monedas_disponibles = [1, 2, 5, 10, 20, 50]
monto = 96

init = time.time()

resultado = cambio_de_moneda_voraz(monedas_disponibles, monto)
print(resultado)
print(f'Tiempo de cálculo {time.time() - init:.15f}s')
```

## Opción 1: Implementa una solución al problema de las 8 reinas en un tablero 8x8

```python
def imprimir_tablero(tablero):
    for i in range(8):
        for j in range(8):
            if tablero[i] == j:
                print("♛", end="")
            else:
                print(". ", end="")
        print()

def es_viable(tablero, fila, columna):
    # Verifica si hay una reina en la columna
    for i in range(fila):
        if tablero[i] == columna:
            return False

    # Verifica si hay una reina en la diagonal superior izquierda
    i = fila - 1
    j = columna - 1
    while i >= 0 and j >= 0:
        if tablero[i] == j:
            return False
        i -= 1
        j -= 1

    # Verifica si hay una reina en la diagonal superior derecha
    i = fila - 1
    j = columna + 1
    while i >= 0 and j < 8:
        if tablero[i] == j:
            return False
        i -= 1
        j += 1

    return True

def resolver_8_reinas(tablero, fila):
    if fila == 8:
        # Caso en el que se colocan todas las reinas correctamente
        imprimir_tablero(tablero)
        return True

    for columna in range(8):
        if es_viable(tablero, fila, columna):
            tablero[fila] = columna
            if resolver_8_reinas(tablero, fila + 1):
                return True
            tablero[fila] = -1

    return False
```

```python
# Resolución del problema de las 8 reinas con una única solución
import numpy as np

# Inicializamos el tablero
tablero = [-1, -1, -1, -1, -1, -1, -1, -1,]

resolver_8_reinas(tablero, 0)
```

## Opción 2: Implementa una solución del problema de las Torres de Hanoi para un número arbitrario n de fichas

```python
# Torres de Hanoy
# Escribe aquí tu código
def hanoi(n, origen, auxiliar, destino):
    if n > 1:
        hanoi(n - 1, origen, destino, auxiliar)
        print(f"Mueve el disco {n} de {origen} a {destino}")
        hanoi(n - 1, auxiliar, origen, destino)

    else:
        print(f"Mueve el disco 1 de {origen} a {destino}")
        return
```

```python
# Probamos el codigo
# Ejemplo usando 10 discos
# Los discos estan nombrados numericamente, donde el valor mas
# pequeño es el disco mas pequeño de tamaño
num_disc = 10

init = time.time()

hanoi(num_disc, "A", "B", "C")
print(f'Tiempo de cálculo {time.time() - init:.15f}s')
```
