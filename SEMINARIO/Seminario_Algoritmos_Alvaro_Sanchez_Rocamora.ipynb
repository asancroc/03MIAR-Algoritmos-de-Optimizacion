## Problema 3. Combinar cifras y operaciones

- El problema consiste en analizar el siguiente problema y diseñar un algoritmo que lo resuelva.
- Disponemos de las 9 cifras del 1 al 9 (excluimos el cero) y de los 4 signos básicos de las
operaciones fundamentales: suma(+), resta(-), multiplicación(*) y división(/)

- Debemos combinarlos alternativamente sin repetir ninguno de ellos para obtener una
cantidad dada. Un ejemplo sería para obtener el 4:
4+2-6/3*1 = 4

- Debe analizarse el problema para encontrar todos los valores enteros posibles planteando las
siguientes cuestiones:
    - ¿Qué valor máximo y mínimo se pueden obtener según las condiciones del problema?
    - ¿Es posible encontrar todos los valores enteros posibles entre dicho mínimo y máximo ?

```python
import itertools
from typing import List, Tuple, Set
```

```python
def combinar_numeros_y_operadores(
    cifras: List[int], 
    operaciones: List[str],
    cant_objetivo: int
    ) -> Tuple[Set[str], float, float]:
    
    valor_min = 100000
    valor_max = -100000

    s_combinaciones = set()

    # Generamos todas las permutaciones de cifras y operaciones
    # teniendo en cuenta las restricciones del problema.
    # Cogemos todas las posiblemas combinaciones de 5 cifras
    for cifras_permutation in itertools.permutations(cifras, 5):
        # Cogemos todas las posiblemas combinaciones posibles de las 4 operaciones
        for operaciones_permutation in itertools.permutations(operaciones, 4):
            expression = ''
            
            # Mergeamos ambas listas para tener todas las combinaciones de num y cifras
            l_expresion = [val for pair in zip(cifras_permutation, operaciones_permutation) for val in pair]
            l_expresion += str(cifras_permutation[-1])

            expression = ''.join(str(val) for val in l_expresion)

            result = eval(expression)

            if result == cant_objetivo:
                s_combinaciones.add(expression)

            valor_min = min(valor_min, result)
            valor_max = max(valor_max, result)

    return s_combinaciones, valor_min, valor_max
```

```python
cifras = [1, 2, 3, 4, 5, 6, 7, 8, 9]
operaciones = ['+', '-', '*', '/']
cant_objetivo = 4

combinaciones, v_minimo, v_maximo = combinar_numeros_y_operadores(
    cifras,
    operaciones,
    cant_objetivo
)

print(f"Numero de expresiones encontradas cuyo resultado es {cant_objetivo}: {len(combinaciones)}")
print(f"El valor más pequeño obtenido: {v_minimo}")
print(f"El valor más grande obtenido: {v_maximo}")
print("Las combinaciones resultantes:")
for expresion in combinaciones:
    print(expresion)
```

3. Flexibilidad para la adición y eliminación de elementos: Un conjunto nos permite agregar nuevas combinaciones válidas de manera rápida y también eliminar elementos si se considera necesario.

Aunque inicialmente una lista podría ser considerada como una opción viable para almacenar las cifras y operadores, el uso de un conjunto se adapta mejor al problema debido a su capacidad para manejar de manera eficiente la eliminación de duplicados y la búsqueda rápida de elementos.

El algoritmo utiliza dos bucles anidados para recorrer todas las permutaciones de cifras y operaciones. Dentro de estos bucles, se genera la expresión y se evalúa utilizando la función eval.

Dado que el algoritmo recorre todas las permutaciones posibles, la complejidad del algoritmo por fuerza bruta se puede estimar como O(9! * 4!) = O(n!).

Como el tamaño del espacio de búsqueda es 9! * 4! es constante, podemos simplificar la complejidad a O(1) ya que el tamaño del espacio de búsqueda no depende del tamaño de entrada.

### **Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta**
Es posible diseñar un algoritmo más eficiente que supere la complejidad temporal del enfoque de fuerza bruta en algunos casos.

El enfoque de fuerza bruta evalúa todas las combinaciones posibles, lo que implica una complejidad factorial. Sin embargo, en ciertos problemas, es posible aplicar técnicas de optimización o algoritmos más avanzados para reducir la complejidad temporal.

Por ejemplo, en el problema de combinar cifras y operaciones para obtener un resultado específico, se puede utilizar la programación dinámica o técnicas de búsqueda en espacio de estados para reducir la complejidad temporal.

La programación dinámica puede aprovechar subproblemas superpuestos y memorización para evitar recalcular soluciones ya conocidas, reduciendo así la cantidad total de cálculos requeridos. Esto puede conducir a una complejidad temporal más eficiente en comparación con el enfoque de fuerza bruta.

```python
def combinar_numeros_y_operadores_dinam(cifras, operaciones, cant_objetivo):
    resultados = []

    # Creamos una matriz para almacenar los resultados
    dp = [[[] for _ in range(cant_objetivo + 1)] for _ in range(len(cifras) + 1)]

    # Recorremos la matriz
    for i in range(1, len(cifras) + 1):
        for j in range(1, cant_objetivo + 1):
            # Aplicamos las operaciones a las subexpresiones
            for operacion in operaciones:
                if operacion == '+' and j >= cifras[i - 1]:
                    for exp in dp[i][j - cifras[i - 1]]:
                        dp[i][j].append(exp + '+' + str(cifras[i - 1]))
                elif operacion == '-' and j + cifras[i - 1] <= cant_objetivo:
                    for exp in dp[i][j + cifras[i - 1]]:
                        dp[i][j].append(exp + '-' + str(cifras[i - 1]))
                elif operacion == '*' and j % cifras[i - 1] == 0 and j // cifras[i - 1] <= cant_objetivo:
                    for exp in dp[i][j // cifras[i - 1]]:
                        dp[i][j].append(exp + '*' + str(cifras[i - 1]))
                elif operacion == '/' and cifras[i - 1] != 0 and j * cifras[i - 1] <= cant_objetivo:
                    for exp in dp[i][j * cifras[i - 1]]:
                        dp[i][j].append(exp + '/' + str(cifras[i - 1]))

    # Obtenemos los resultados finales de la última celda de la matriz
    resultados = dp[len(cifras)][cant_objetivo]

    return resultados
```

```python
cifras = [1, 2, 3, 4, 5, 6, 7, 8, 9]
operaciones = ['+', '-', '*', '/']
resultado_objetivo = 4

resultados = combinar_numeros_y_operadores_dinam(cifras, operaciones, resultado_objetivo)
print(resultados)
```
